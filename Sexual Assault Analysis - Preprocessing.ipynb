## Read  files from 2010 to 2016 ##

```python
datypes = {'FECHA':str}
parse_dates = ['FECHA']
names =["FECHA","DEPARTAMENTO ","MUNICIPIO","DIA ","HORA","BARRIO","ZONA","CLASE SITIO","DESCRIPCION CONDUCTA",
        "ARMA EMPLEADA","MOVIL AGRESOR","MOVIL VICTIMA","EDAD","SEXO","ESTADO CIVIL","PAIS NACE","CLASE EMPLEADO",
        "PROFESION","ESCOLARIDAD","CODIGO DANE","YEAR"]
names_mod =["FECHA","DEPARTAMENTO ","MUNICIPIO","DESCRIPCION CONDUCTA","DIA ","HORA","BARRIO","ZONA","CLASE SITIO",
        "ARMA EMPLEADA","MOVIL AGRESOR","MOVIL VICTIMA","EDAD","SEXO","ESTADO CIVIL","PAIS NACE","CLASE EMPLEADO",
        "PROFESION","ESCOLARIDAD","CODIGO DANE","YEAR"]

year10 = pd.read_csv("Delito_Sexuales_2010.csv", sep=",",dtype=datypes, parse_dates=parse_dates,dayfirst=True,encoding='utf-8',header=0, names=names)
year11 = pd.read_csv("Delito_Sexuales_2011.csv", sep=",",dtype=datypes, parse_dates=parse_dates,dayfirst=True,encoding='utf-8',header=0, names=names)
year12 = pd.read_csv("Delito_Sexuales_2012.csv", sep=",",dtype=datypes, parse_dates=parse_dates,dayfirst=True,encoding='utf-8',header=0, names=names)
year13 = pd.read_csv("Delito_Sexuales_2013.csv", sep=",",dtype=datypes, parse_dates=parse_dates,dayfirst=True,encoding='utf-8',header=0, names=names)
year14 = pd.read_csv("Delito_Sexuales_2014.csv", sep=",",dtype=datypes, parse_dates=parse_dates,dayfirst=True,encoding='utf-8',header=0, names=names)
year15 = pd.read_csv("Delito_Sexuales_2015.csv", sep=",",dtype=datypes, parse_dates=parse_dates,dayfirst=True,encoding='utf-8',header=0, names=names)
year16 = pd.read_csv("Delito_Sexuales_2016.csv",  sep=",",dtype=datypes, parse_dates=parse_dates,dayfirst=True,encoding='utf-8',header=0, names=names_mod)
year17 = pd.read_csv("Delito_Sexuales_2017.csv",  sep=",",dtype=datypes, parse_dates=parse_dates,dayfirst=True,encoding='utf-8',header=0, names=names_mod)
```

```python
result = year17.append([year16,year15,year14,year13,year12,year11,year10]);
result.drop('YEAR', axis=1, inplace=True);
```

```python
pandas_profiling.ProfileReport(result)
```

```python
# Clean the SEXO column
result['SEXO'] = result['SEXO'].replace('-',np.nan)
result['SEXO'] = result['SEXO'].replace('NO REPORTADO',np.nan)
result['SEXO'] = result['SEXO'].replace('NO REPORTA',np.nan)

# Clean the ARMA EMPLEADA column
result['ARMA EMPLEADA'] = result['ARMA EMPLEADA'].replace('-',np.nan)
result['ARMA EMPLEADA'] = result['ARMA EMPLEADA'].replace('NO REPORTADO',np.nan)
result['ARMA EMPLEADA'] = result['ARMA EMPLEADA'].replace('ARMA BLANCA / CORTOPUNZANTE','ARMA BLANCA')
result['ARMA EMPLEADA'] = result['ARMA EMPLEADA'].replace('CORTOPUNZANTES','ARMA BLANCA')
result['ARMA EMPLEADA'] = result['ARMA EMPLEADA'].replace('CORTANTES','ARMA BLANCA')
result['ARMA EMPLEADA'] = result['ARMA EMPLEADA'].replace('ARMAS BLANCAS','ARMA BLANCA')
result['ARMA EMPLEADA'] = result['ARMA EMPLEADA'].replace('CINTAS','CINTAS/CINTURON')

# Clean the CLASE EMPLEADO column
result['CLASE EMPLEADO'] = result['CLASE EMPLEADO'].replace('NO REPORTA',np.nan)
result['CLASE EMPLEADO'] = result['CLASE EMPLEADO'].replace('NO REPORTADO',np.nan)

# Clean the EDAD EMPLEADO column
result['EDAD'] = result['EDAD'].replace('NO REPORTADO',np.nan)
result['EDAD'] = result['EDAD'].replace('NO REPORTADO1',np.nan)

# Clean the ESCOLARIDAD column
result['ESCOLARIDAD'] = result['ESCOLARIDAD'].replace('NO REPORTADO',np.nan)
result['ESCOLARIDAD'] = result['ESCOLARIDAD'].replace('-',np.nan)

# Clean the ESTADO CIVIL column
result['ESTADO CIVIL'] = result['ESTADO CIVIL'].replace('NO REPORTADO',np.nan)
result['ESTADO CIVIL'] = result['ESTADO CIVIL'].replace('NO REPORTA',np.nan)
result['ESTADO CIVIL'] = result['ESTADO CIVIL'].replace('-',np.nan)

# Clean the MOVIL AGRESOR column
result['MOVIL AGRESOR'] = result['MOVIL AGRESOR'].replace('-',np.nan)
result['MOVIL AGRESOR'] = result['MOVIL AGRESOR'].replace('NO REPORTADO',np.nan)
result['MOVIL AGRESOR'] = result['MOVIL AGRESOR'].replace('VEHICULO','CONDUCTOR VEHICULO')

# Clean the MOVIL VICTIMA column
result['MOVIL VICTIMA'] = result['MOVIL VICTIMA'].replace('-',np.nan)
result['MOVIL VICTIMA'] = result['MOVIL VICTIMA'].replace('NO REPORTADO',np.nan)
result['MOVIL VICTIMA'] = result['MOVIL VICTIMA'].replace('VEHICULO','CONDUCTOR VEHICULO')

# Clean the PAIS NACE column
result['PAIS NACE'] = result['PAIS NACE'].replace('-',np.nan)
result['PAIS NACE'] = result['PAIS NACE'].replace('NO REPORTADO',np.nan)
```

```python
# We change the hour format, only preserving the hour data, droping minutes and seconds
for x in range(0, 24):
    result.loc[result.HORA.str.startswith(str(x) + ':',na=False),'HORA'] = x
```

```python
# We remove the Codigo Dane column as it is not adding info
result.drop('CODIGO DANE', axis=1, inplace=True);
```

```python
# We save the pandas dataframe as a new csv file to make further analysis
result.to_csv('sexual_offenses.csv', encoding='utf-8')
```
